# seaborn - Statistical data visualisation Library
## What is seaborn?
seaborn is a python library that builts on the top of matplotlib.That makes the statistical data more easy and beautiful.
## Why we need it?
1. For creating plots like violin plot,box plot, and pair plots.
2. Automatically handles data from Pandas DataFrames.
## Where we use it ?
1. DataAnlaysis and reporting.
2. Exploring relationship between variables.
3. Machine Learning (For Visualising datasets and results).


```python
#2. Importing seaborn
import seaborn as sns
```

```python
import sys
print(sys.executable)
```

```python
# Load Titanic dataset
df=sns.load_dataset("titanic")
df.head()
```

```python
df.tail()
print("Shape :",df.shape)
print(df.info())
print(df.describe())
```

```python
## 4. Histograms & Countplots
# Histogram
import matplotlib.pyplot as plt
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
 
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
 
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
 
```

```python
# "Show trends or changes in mean values across categories."
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
```

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
```

1. Compare Passenger Survival by Class and Gender
2. Distribution of Passenger Ages
3. Fare Distribution by Class
4. Relationship Between Age and Fare

```python
# 1. Compare Passenger Survival by Class and Gender
sns.barplot(x='class', y='survived', hue='sex', data=df, ci=None, palette='Set1')
plt.title('Survival Rate by Class and Gender')
plt.ylabel('Survival Rate')
plt.show()

# 2. Distribution of Passenger Ages
sns.histplot(df['age'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.show()

# 3. Fare Distribution by Class
sns.boxplot(x='class', y='fare', data=df, palette='pastel')
plt.title('Fare Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()

# 4. Relationship Between Age and Fare
sns.scatterplot(x='age', y='fare', data=df, hue='sex', style='class', palette='Set1')
plt.title('Relationship Between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# 5. Correlation Heatmap
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Titanic Dataset)')
plt.show()
```
