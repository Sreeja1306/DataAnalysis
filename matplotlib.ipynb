## Matplotlib - Python Visualisation Library

# 1. What is matplotlib?
matplotlib is python library for creating static animated and interactive vizualisations .
# 2. Why we need it ?
It helps to understand trends,patterns and outlayers.
# 3. Where we use it?
1. In data Analysis.
2. In Machine Learning (For plotting loss curves,Linear Regression ,Logistic Regression).
3. For Buisness Purposes(Like to analyze the market trends,Finance and Scientific needs)

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
 
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.bar(categories, values, color='green')
plt.title("Bar Plot Example")
plt.show()
```

```python
x = [5,7,8,7,2,17,2,9,4,11]
y = [99,86,87,88,100,86,103,87,94,78]
 
plt.scatter(x, y, color='red', marker='x')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
data = [22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
 
plt.hist(data, bins=6, color='orange', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
 
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Line 1')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
 
plt.subplot(1,2,1)  # 1 row, 2 columns, plot 1
plt.plot(x, y1, color='blue')
plt.title("Plot 1")
 
plt.subplot(1,2,2)  # 1 row, 2 columns, plot 2
plt.plot(x, y2, color='red')
plt.title("Plot 2")
 
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.cos(x)
plt.stem(x, y)
plt.title("Stem Plot Example")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,5,6,8]
plt.step(x, y, where='mid', color='magenta')
plt.title("Step Plot Example")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.fill_between(x, y, color='lightblue', alpha=0.6)
plt.title("Area Plot Example")
plt.show()
```

```python
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
 
plt.contour(X, Y, Z, cmap='plasma')
plt.title("Contour Plot Example")
plt.colorbar()
plt.show()
```

```python
plt.contourf(X, Y, Z, cmap='viridis')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(2*theta))
plt.polar(theta, r)
plt.title("Polar Plot Example")
plt.show()
```

```python
x = np.arange(1,6)
y = np.random.randint(5,15,5)
errors = np.random.rand(5)
 
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Plot Example")
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Plot")
plt.show()
 
```

```python
from mpl_toolkits.mplot3d import Axes3D
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title("3D Surface Plot")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t
 
ax.plot3D(x, y, z, 'purple')
ax.set_title("3D Line Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue')
ax.set_title("3D Sphere Surface")
plt.show()
```

```python
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)
ax.set_title("3D Surface + Wireframe")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=5, cmap='coolwarm')
 
ax.set_title("3D Contour Projection on Planes")
plt.show()
 
```

3.Make a scatter plot of Age vs Salary.
 
4.Draw a histogram of marks scored by students in a class.

6.Create two subplots: Line plot and bar plot side by side.

```python
# 1.Create a line plot of student marks across 5 exams.
x = ["java","python","DED","DM","C"]
y = [10,20,25,30,40]
plt.plot(x, y)
plt.title("line plot of student marks across 5 exams")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

# 2.Plot a bar chart of sales of 4 products.
categories = ['Product A','Product B','Product C','product D']
values = [10,15,7,12]
plt.bar(categories, values, color='lightblue')
plt.title("Bar chart of sales of 4 products")
plt.show()

# 3.Make a scatter plot of Age vs Salary.
Age = [25,32,27,42,29,34,31]
Salary = [20000,32000,28000,45000,30000,35000,33000]
plt.scatter(Age, Salary, color='red', marker='.')
plt.title("Scatter plot of Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

# 4.Draw a histogram of marks scored by students in a class.
marks = [87,56,73,55,64,60,79]
plt.hist(marks, bins=6, color='orange', edgecolor='black')
plt.title("Histogram of marks scored by students in a class")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()

# 5.Customize a plot with different colors, markers, and line styles.
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, color='red', linestyle='--', marker='s', label='Line 1')
plt.plot(x, [i*1.5 for i in y], color='blue', linestyle='-.', marker='o', label='Line 2')
plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# 6.Create two subplots: Line plot and bar plot side by side.
import matplotlib.pyplot as plt
exams = ['OS', 'Java',  'C', 'Python']
marks = [78, 85, 90, 88]
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 120, 180]
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(exams, marks, marker='o', color='teal')
axs[0].set_title('Line Plot: Marks')
axs[0].set_xlabel('Exams')
axs[0].set_ylabel('Marks')
axs[1].bar(products, sales, color='coral')
axs[1].set_title('Bar Plot: Sales')
axs[1].set_xlabel('Products')
axs[1].set_ylabel('Sales')
plt.tight_layout()
plt.show()
```
